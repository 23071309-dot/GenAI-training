```python
from langchain_community.document_loaders import TextLoader
loadder_fixed = TextLoader('/content/sample_data/ML.txt', encoding="utf-8")
text_documents = loadder_fixed.load()
text_documents
```

```python
# Text splitter
from langchain_text_splitters import RecursiveCharacterTextSplitter
textSplitter = RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=20)
finalDocuments = textSplitter.split_documents(text_documents)
finalDocuments
```

```python
# If finalDocuments is a list of Document objects:
texts = [doc.page_content for doc in finalDocuments]
```

```python
##Huggingface embedding
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
!pip install faiss-cpu
# vector Store in DB
from langchain_community.vectorstores import FAISS
db = FAISS.from_documents(finalDocuments, embeddings)
db
```

```python
query = input("Enter your question for ML : ")
docs = db.similarity_search(query)
for i in range(len(docs)):
    print(docs[i].page_content)
```
